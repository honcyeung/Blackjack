```python
import random
from IPython.display import clear_output
```

```python
cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]

def draw(): return random.choice(cards)

def calculate_score(cards): 
    
    if sum(cards) == 21 and len(cards) == 2:
        return 0
    if 11 in cards and sum(cards) > 21:
        cards.remove(11)
        cards.append(1)
    return sum(cards)

def compare(player_score, computer_score):
    if player_score == computer_score:
        return 'Draw!'
    elif computer_score == 0:
        return 'Lost, computer has Blackjack!'
    elif player_score == 0:
        return 'Win with Blackjack!'
    elif player_score > 21:
        return 'You lost!'
    elif computer_score > 21:
        return 'Computer lost!'
    elif computer_score > player_score:
        return 'You lost!'
    elif player_score > computer_score:
        return 'You won!'
    else:
        return 'You lost!'

def blackjack():

    if restart == 0:
        decision = input('Do you want to play Blackjack (y/n)? ').lower()
    else:
        decision = 'y'

    if decision == 'y':
        player, computer = [], []
        is_game_over = False

        for _ in range(2): 
            player.append(draw())
            computer.append(draw())

        while not is_game_over:
            player_score = calculate_score(player)
            computer_score = calculate_score(computer)
            print(f'\nYour cards: {player}, current score: {player_score}')
            print(f"Computer's first card: {computer[0]}")

            if player_score == 0 or computer_score == 0 or player_score > 21:
                is_game_over = True
            else:
                another = input("Type 'y' to get another card, type 'n' to pass: ")
                if another == 'y':
                    player.append(draw())
                else:
                    is_game_over = True

        while computer_score != 0 and computer_score < 17:
            computer.append(draw())
            computer_score = calculate_score(computer)

        print(f'Your final hand: {player}, score: {player_score}')
        print(f"Computer's final hand: {computer}, score: {computer_score}")
        print(compare(player_score, computer_score))
        
    else:
        print('End.')

restart = 0
blackjack()

while input('Do you want to start again (y/n)? ').lower() == 'y':
    clear_output(wait = False)
    restart = 1
    blackjack()
else:
    print('End.')
```
